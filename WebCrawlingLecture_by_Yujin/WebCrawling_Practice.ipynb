```
# 초기 환경 설정 및 크롤링
from urllib.request import urlopen
import bs4

url = "https://finance.naver.com/"
html = urlopen(url)

bs_obj = bs4.BeautifulSoup(html, "html.parser")

print(bs_obj)
```

```
"""
우리가 찾고자 하는 기사 제목 위치
div id = wrap > div id = container > div id = content > div class = article > div class = section > div class = news_area
> div class = section_strategy > ul > li > span > a
"""

# section_strategy div 추출
section_strategy = bs_obj.find("div", {"class":"section_strategy"})
print(section_strategy)
```

```
# findAll 함수로 li 태그 추출
list1 = section_strategy.findAll("li")

for li in list1:
  print(li)
```

```
print(list1)
```

```
# find 함수로 span 태그 추출
for i in range(0, len(list1)):
  list1[i] = list1[i].find("span")

  print(list1[i])
```

```
# find 함수로 a 태그 추출
for i in range(0, len(list1)):
  list1[i] = list1[i].find("a")

  print(list1[i])
```

```
# a 태그 안의 텍스트 출력
for i in range(0, len(list1)):
  print(list1[i].text)
```
