```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
req=requests.get('https://finance.naver.com/news/news_search.nhn?rcdate=&q=%C4%DA%B7%CE%B3%AA&x=0&y=0&sm=all.basic&pd=4&stDateStart=2020-02-20&stDateEnd=2020-03-04')
html=req.content
soup=BeautifulSoup(html,'html.parser',from_encoding='ecu-kr')
```

```python
s_num=str(soup.select('.pgRR')[0]).find('page')
page=int(str(soup.select('.pgRR')[0])[s_num+5:s_num+9])
```

```python
link='https://finance.naver.com/news/news_search.nhn?rcdate=&q=%C4%DA%B7%CE%B3%AA&x=0&y=0&sm=all.basic&pd=4&stDateStart=2020-02-20&stDateEnd=2020-03-04&page='
result=[]
for i in range(1,page,1):
    new=link+str(i)
    req=requests.get(new)
    html=req.content
    soup=BeautifulSoup(html,'html.parser',from_encoding='ecu-kr')
    lst=soup.select('.articleSubject')
    result.extend(lst)
    
```

```python
for el in result:
    print(el.get_text().replace('\n',''))
```
